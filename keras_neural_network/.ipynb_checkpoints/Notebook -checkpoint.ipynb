# Red neuronal modelado de datos

En este notebook vamos a realziar el desarrollo de la red neuronal que utilizaremos para predecir el precio de los pisos en función a distintas variables

Primero que nada, importamos todas las librerías necesarias.
En este caso vamos a trabajar con gran cantidad de funciones que provienen de la librería Keras. Keras es una librería que utiliza backend de TensorFlow.

```python
from numpy import loadtxt
from keras.models import Sequential
from keras.layers import Dense
import keras as keras
from io import open
import csv
import numpy as np
import pandas
import sys
import matplotlib.pyplot as plt
```

Cargamos el dataset con el que vamos a trabajar 

```python
np.set_printoptions(threshold=sys.maxsize)

csv_route = 'finalDataset3.csv'
```

Generamos la red neuronal
En este punto establecemos el número de entradas y salidas de nuestra red, en nuestro caso seis entradas y un valor de salida.
Trabajamos con una capa intermedia, en el caso de querer trabajar con más deberíamos añadirselas al modelo (Ej:model.add(Dense(6, activation='relu'))

```python
dataNaN = pandas.read_csv(csv_route, header=0, encoding='latin1')
data = dataNaN.fillna(value=1).drop(['Tipo','Distrito'],axis=1).to_numpy()

X = data[:,2:8]
Y = data[:,1]

model = Sequential()

model.add(Dense(6, input_dim=6, activation='relu'))

model.add(Dense(1, activation='relu'))
```

Compilamos el modelo. 
En este paso podemos modificar el optimizador a utilizar (SGD, RMSprop o Adam) así como el Learning Rate(lr).

Ej: keras.optimizers.OPTIMIZADOR(VALOR_LR)

También podemos modificar la función de regresión utilizada para las pérdidas(mean_squared_logarithmic_error, huber o log_cosh)
Bastaría con modificar el campo loss. 
Ej: loss='FUNCION_PERDIDAS'

```python
optimizer = keras.optimizers.SGD(lr=0.1)
model.compile(loss='mean_squared_logarithmic_error', optimizer=optimizer, metrics=['msle'])
```

Entrenamos la red.
Modificando el valor de el campo "epochs" modificamos el número de épocas por las que pasará nuestra red en su entrenamiento.

```python
history = model.fit(X, Y, validation_split=0.30, epochs=150, batch_size=120)
print('Minimo msle: ',np.amin(history.history['msle']))
```
